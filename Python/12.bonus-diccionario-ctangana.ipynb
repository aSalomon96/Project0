***HINT***: Ordenar los valores de un diccionario es más complicado que ordenar las claves porque los valores de un diccionario no son únicos. Por lo tanto, no puedes usar la misma forma en que ordenaste las claves del diccionario para ordenar los valores del diccionario. La forma de ordenar un diccionario por valor es utilizar las funciones sorted y operator.itemgetter. 


1. Busca la letra de C. Tangana, Andrés Calamaro - Hong Kong y almacenala como una cadena de strings en `letra_cancion`.

```python
letra_cancion = """
Voy a salir, no sé si vuelva
Voy a salir de dudas y deudas
Con todo lo que no he dicho
Y con lo que no diré nunca
Voy a cerrar, ¿quién dijo miedo?
No tengo nada de lo que era
Tampoco espero lo que quieras
El que no arriesga, no gana nada

He bajado al infierno
Por una mujer, por una mujer
He bebido en exceso
Por una mujer, por una mujer

Me cago en el dinero
Me cago en el amor
Me cago en tu mareo
Me cago en el dolor
Me cago en la fama
Me cago en el poder
Me cago en tus amigos
Y en todo lo que sos
Hong Kong, Hong Kong, Hong Kong

Voy a salir, no sé si vuelva
Voy a salir de dudas y deudas
Con todo lo que no he dicho
Y con lo que no diré nunca

Voy a cerrar, ¿quién dijo miedo?
No tengo nada de lo que era
Tampoco espero lo que quieras
El que no arriesga, no gana nada

He bajado al infierno
Por una mujer, por una mujer
He bebido en exceso
Por una mujer, por una mujer

Me cago en el dinero
Me cago en el amor
Me cago en tu mareo
Me cago en el dolor
Me cago en la fama
Me cago en el poder
Me cago en tus amigos
Y en todo lo que sos
Hong Kong, Hong Kong, Hong Kong

Yo he bajado al infierno
Por una mujer, por una mujer
He bebido en exceso
Por una mujer, por una mujer
Hong Kong, Hong Kong
Hong Kong, Hong Kong, Hong Kong
Hong Kong, Hong Kong
Hong Kong, Hong Kong, Hong Kong
Hong Kong, Hong Kong, Hong Kong
"""

```

2. Crea una lista `palabras` utilizando `letra_cancion`.

```python
palabras = letra_cancion.split()
```

3. Crea un diccionario `frecuencia_palabras` en el que aparezca como claves cada una de las palabras de `letra_cancion` y sus valores que sea el numero de ocurrencia

```python
from collections import Counter

# Crear el diccionario de frecuencia de palabras
frecuencia_palabras = dict(Counter(palabras))

# Mostrar el diccionario
frecuencia_palabras
```

4. Ordena alfabeticamente las claves de `frecuencia_palabras` 

```python
# Obtener las claves ordenadas alfabéticamente
claves_ordenadas = sorted(frecuencia_palabras.keys())

# Mostrar las claves ordenadas
claves_ordenadas
```

5. Crea un nuevo diccionario llamado `frecuencia_palabras2` con el orden obtenido en la pregunta 4.

```python
# Crear el nuevo diccionario con las claves ordenadas alfabéticamente
frecuencia_palabras2 = {clave: frecuencia_palabras[clave] for clave in claves_ordenadas}

# Mostrar el nuevo diccionario
frecuencia_palabras2
```

6. Ordenar los valores de `frecuencia_palabras2` de forma ascendente.


```python
# Ordenar el diccionario por los valores (frecuencia) de forma ascendente
frecuencia_palabras_ordenado_por_valor = dict(sorted(frecuencia_palabras2.items(), key=lambda item: item[1]))

# Mostrar el diccionario ordenado por valores
frecuencia_palabras_ordenado_por_valor
```


***HINT***: Ordenar los valores de un diccionario es más complicado que ordenar las claves porque los valores de un diccionario no son únicos. Por lo tanto, no puedes usar la misma forma en que ordenaste las claves del diccionario para ordenar los valores del diccionario. La forma de ordenar un diccionario por valor es utilizar las funciones sorted y operator.itemgetter. 

```python
import operator

# Ordenar el diccionario `frecuencia_palabras2` por valores de forma ascendente
frecuencia_palabras_ordenado_por_valor = dict(sorted(frecuencia_palabras2.items(), key=operator.itemgetter(1)))

# Mostrar el diccionario ordenado por valores
frecuencia_palabras_ordenado_por_valor
```
